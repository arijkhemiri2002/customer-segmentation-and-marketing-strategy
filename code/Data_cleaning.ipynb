## Data Cleaning & Preprocessing

```python
import pandas as pd
data = pd.read_csv("../data/marketing_campaign.csv", sep="\t")
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
#faire un copie de data pour ne pas toucher à l'original
copie=data.copy()
```

# Suppression des valeurs nulles

```python
#d'apres les infos il y a des valeurs manquantes dans la colonne "Income" d'ou on va les supprimer car il sont trés negligables
copie = copie.dropna(subset=["Income"])
```

```python
#longeur du dataframe après suppression des valeurs manquantes
print(len(copie))
```

# Suppression des colonnes inutiles

```python
#d'aprés le head j'ai remarqué que les colonnes "AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5","Complain","Response" contenaient bcp de valeurs nulles d'ou je vais les supprimer
copie["AcceptedCmp1"].value_counts()[0]
copie["Complain"].value_counts()[0]
copie["Response"].value_counts()[0]
copie = copie.drop(columns=["AcceptedCmp1","AcceptedCmp2","AcceptedCmp3","AcceptedCmp4","AcceptedCmp5","Complain","Response"])
```

```python
#j'ai constater que les colonnes "Z_Revenue" et "Z_CostContact" contiennent la meme veleur ds toutes les lignes d'ou je vais les supprimer
copie.Z_Revenue.value_counts()
copie.Z_CostContact.value_counts()
copie.drop(columns=["Z_CostContact","Z_Revenue"], inplace=True)
```

```python
copie.head()
```

# Encodage des valeurs categorielle

```python
#mainteant je vais faire le traitement de la colonne "Education" pour la convertir en numérique
copie.Education.value_counts() #j'ai commencer par voir les différentes catégories de la colonne "Education"
```

```python
#maintenant pour les valeur categorielle je vais faire un ordianl encoding pour les transformer en valeur numerique (plus significatif))
mapping = {"Basic":1, "2n Cycle":2, "Graduation":3, "Master":4, "PhD":5}
copie["Education_encoded"] = copie["Education"].map(mapping)
```

```python
#meme chose pour la colonne "Marital_Status"
copie.Marital_Status.value_counts()
```

```python
copie["Marital_Status"] = copie["Marital_Status"].replace(["YOLO","Absurd","alone"], "Other")
```

```python
mapping = {"Single":1, "Together":2, "Married":3, "Divorced":4, "Widow":5, "Other":0}
copie["Marital_Status_encoded"] = copie["Marital_Status"].map(mapping)
```

# Feature engineering

```python
from datetime import datetime
#d'apres la colonne "Year_Birth" on peut calculer l'age de chaque client pour avoir une meilleure idée sur la distribution d'age dans notre dataset
current_year = datetime.now().year
copie["Age"] = current_year - copie["Year_Birth"]

print(copie[["Year_Birth", "Age"]].head())
```

```python
#maintenant on peut faire le meme chose pour la colonne "Dt_Customer" pour calculer le nombre d'année d'enrollement de chaque client
copie["Dt_Customer"] = pd.to_datetime(copie["Dt_Customer"], format="%d-%m-%Y")

copie['Seniority'] = (pd.to_datetime('today') - copie["Dt_Customer"]).dt.days // 30
```

```python
#c'est la somme de depense de chaque client
copie['TotalMnt'] = copie[['MntWines','MntFruits','MntMeatProducts',
                      'MntFishProducts','MntSweetProducts','MntGoldProds']].sum(axis=1)
```

```python
#permet d'analyser les paniers d'achat de chaque client
copie['TotalPurchases'] = copie[['NumWebPurchases','NumCatalogPurchases',
                            'NumStorePurchases','NumDealsPurchases']].sum(axis=1)
```

```python
#represente la charge familiale réelle du client
copie['TotalChildren'] = copie['Kidhome'] + copie['Teenhome']
```

```python
#c'est la Part du revenu consacrée aux achats
copie['SpendPerIncome'] = copie['TotalMnt'] / (copie['Income'] + 1)
```

```python
#j'ai supprimer les colonnes "Year_Birth","Dt_Customer","Education","Marital_Status","ID" car elles sont soit redondantes ou pas utiles pour l'analyse
copie.drop(columns=["Year_Birth","Dt_Customer","Education","Marital_Status","ID","Kidhome","Teenhome","MntWines","MntFruits","MntMeatProducts",
                      "MntFishProducts","MntSweetProducts","MntGoldProds","NumWebPurchases","NumCatalogPurchases",
                            "NumStorePurchases","NumDealsPurchases"], inplace=True)
```

```python
copie.head()
copie= copie.dropna()
```

```python
#maintenat tous les données sont numerique d'ou on va faire standardisation pour les mettre à la même échelle
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
copie_scaled = scaler.fit_transform(copie)
```

```python
#aprés scaling c'est necessaire de verifier les valeurs nan
import numpy as np
print(np.isnan(copie_scaled).sum())
```

```python
#resultats aprés nettoyage et standardisation
copie_scaled=pd.DataFrame(copie_scaled, columns=copie.columns)
copie_scaled.head()
```
